```python
import pandas as pd
import ast
```

```python
df = pd.read_csv("eurocup_2020_results.csv")
```

```python
df.head() #Reviso los datos
```

```python
def read_d_data(elemento):
    try:
        return ast.literal_eval(elemento)  #Hago una función para convertir un tipo de dato en el dato que "simula" ser.
    except:
        return False
```

```python
df["lineup_home"] = df["lineup_home"].apply(read_d_data) #Convierto estos tres STR en listas.
df["lineup_away"] = df["lineup_away"].apply(read_d_data)
df["events_list"] = df["events_list"].apply(read_d_data)
```

```python
def players(lista):
    team = []
    player = ""
    number = ""
    for i in range(11):
        player = lista[i]["Player_Name"]
        number = lista[i]["Player_Number"]
        team.append(f"{player} ({number})")
    return team
```

```python
df["lineup_home"] = df["lineup_home"].apply(players)
```

```python
df["lineup_away"] = df["lineup_away"].apply(players)
```
